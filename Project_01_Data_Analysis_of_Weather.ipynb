# **Temperature Trends**


```python
df['Data/Time']=pd.to_datetime(df['Date/Time'])
plt.plot(df['Date/Time'],df['Temp_C'])
plt.title('Temperature Trends Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature(Celcius)')
plt.show()
```
